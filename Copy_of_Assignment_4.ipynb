Use one hot encoding to encode features/ columns like

````
# your answer here to convert sex col to one hot encoding
train_data['Sex'] = train_data['Sex'].astype('category')
train_data = pd.get_dummies(train_data, columns=['Sex'])
train_data.columns
````

````
# your answer here to convert pclass col to one hot encoding
train_data['Pclass'] = train_data['Pclass'].astype('category')
train_data = pd.get_dummies(train_data, columns=['Pclass'])
train_data.columns
````

````
# your answer here to convert Embarked col to one hot encoding
train_data['Embarked'] = train_data['Embarked'].astype('category')
train_data = pd.get_dummies(train_data, columns=['Embarked'])
train_data.columns
````

## Task 6

Splitting train and validation set . This task involves to splitting the original train set to train and validation set.

````
# your answer here
````

````
# your answer here

from sklearn.model_selection import train_test_split

# Separate the input and output columns.
X = train_data.drop('Survived', axis=1)
# For the output, we'll use the Survived column
Y = train_data['Survived']

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2, random_state=0)
````

````
##Verify Data

# summarizes X and Y shape
print(X.shape, Y.shape)
# summarizes Xtrain, Xtest, Ytrain, Ytest
print(Xtrain.shape, Xtest.shape, Ytrain.shape, Ytest.shape)
````

## Task 7

Fitting into machine learning model. Previously in practical we used logistic regression and SVC to classify/predict whether the given passenger would surivive or not. In this **Task 6**, use Decision tree classifer to predict whether the given passenger survives or not. Before fitting the dataset into model make sure that your data is cleaned properly that includes eliminating null data, converting all categorical columns into one hot encoding.

To learn more about Decision tree Classifier : https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html

````
#your answer here

from sklearn import tree
from sklearn.model_selection import cross_val_score

dtc = tree.DecisionTreeClassifier()
dtc.fit(Xtrain, Ytrain)
dtc_acc = cross_val_score(dtc, Xtrain, Ytrain, cv=5, scoring='accuracy').mean()
dtc_recall = cross_val_score(dtc, Xtrain, Ytrain, cv=5, scoring='recall').mean()

print("DecisionTreeClassifier Accuracy =", dtc_acc)
print("DecisionTreeClassifier Recall =", dtc_recall)

````
